### Day 1: Division and Square-root (15-01-2023)

Write a function called divide_or_square that takes one
argument (a number), and returns the square root of the number
if it is divisible by 5, returns its remainder if it is not divisible by 5. For example, if you pass 10 as an argument, then your function
should return 3.16 as the square root.

```python
import math

def divide_or_square(number):
    if number % 5 == 0:
        return f"The sqareroot of {number} is {math.sqrt(number)}"
    else:
        remainder = number % 5
        return f"The remainder of dividing {number} by 5 is {remainder}"
        
print(divide_or_square(10))
print(divide_or_square(8))
```

```python
def longest_value(dict1):
    #Create an empty string for the longest value
    longest = ''
    for value in dict1.values():
        if len(value) > len(longest):          
            longest = value             #This will store the value with highest value of length
    return longest

fruits = {'fruit': 'apple', 'color': 'green'}
longest_value(fruits)
```

Write a function called word_index that takes one argument,
a list of strings and returns the index of the longest word in the
list. If there is no longest word (if all the strings are of the same

below should return 2.

words1 = ["Hate", "remorse", "vengeance"] 

And this list below, shoul return zero (0) 

words2 = ["Love", "Hate"]

```python
def word_index(wordList):
    for word in wordList:
        for i in range(len(wordList) - 1):
            if len(wordList[i]) == len(wordList[i + 1]):
                return 0
            elif len(word) == len(max(wordList)):                                 # Using len function and max to find the longest word
                return f'The longest word is at index: {wordList.index(word)}'    ## using list method index to find, index of the longest word
```

should be separated by dots(.) For example, if you pass 6 as
an argument, your function should return ‘0.1.2.3.4.5’.

```python
def count_Sname(nameList):
    Scount = []
    Sname = {}
    for name in nameList:
        if name[0] == 'S':
            Scount.append(name)
            Sname.update({name: Scount.count(name)})
    return Sname


names = ["Joseph","Nathan", "Sasha", "Kelly", "Muhammad", "Jabulani", "Sera", "Patel", "Sera"]
count_Sname(names)
```

**Extra Challenge: List of Prime Numbers**

Write a function called prime_numbers. This function asks a
user to enter a number (integer) as an argument and returns a
list of all the prime numbers within its range. 

For example, if user
enters 10, your code should return [2, 3, 5, 7] as prime numbers.

```python
def prime_numbers():
    num = int(input("Enter an integer : "))
    prime_num = []
    for i in range(2,num):
        for j in range(2, i):
            if i % j == 0:
                break
        else:
            prime_num.append(i)
    return prime_num

prime_numbers()
```

```python
def prime_numbers() -> list:
    prime_num = []
    n = int(input('Please enter a number (integer): '))
    # Creating a range of the input number remember to add a 1 and the end ensure all numbers in the range are covered
    for i in range(0, n + 1):
           # only numbers above 1 are needed
        if i > 1:
            for j in range(2, i):
                 # if a number in the range is divisible by j. Then number is not prime
                if i % j == 0:
                    break
            else:
                prime_num.append(i)
    return prime_num

print(prime_numbers())
```

**Extra Challenge: Zeros to the End**

Write a function called zeros_last. This function takes a list as
argument. If a list has zeros (0), it will move them to the front of
the list and maintain the order of the other numbers in the list.
If there are no Zeros in the list, the function should return the
original list sorted in ascending order. 

For example, if you pass
[1, 4, 6, 0, 7,0,9] as an argument, your code should return [1,
4, 6, 7, 9, 0, 0]. If you pass [2, 1, 4, 7, 6] as your argument,
your code should return [1, 2, 4, 6, 7].

```python
def zeros_last(arr: list) -> list:
    i = 0 # setting index 0
    arr1 = arr
    for num in arr:
         # Checking for non-zero numbers
        if num != 0:
            # Moving non-zero numbers to the front of the list
            arr[i] = num
            i += 1
                  # Moving zero elements to the end of the list.
    while i < len(arr):
        arr[i] = 0
        i += 1
        return arr
    else:
         # if no zeros return original list in ascending order
        return sorted(arr1)
    
    
list1 = [1, 4, 6, 0, 7, 0]
list2 = [2, 1, 4, 7, 6]
print(zeros_last(list1))
print(zeros_last(list2))
```
